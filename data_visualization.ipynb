```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

```python
with open("reviews.json","r") as f:
    reviews_data = json.load(f)

print(len(reviews_data))
```

```python
flattened_data = []
#for flattening the data of reviews to allow it to be stored in a data frame
for review in reviews_data:
    for key, details in review.items():
        #splitting review number
        review_number = int(key.split("_")[-1])
        #unpacking ratings dict into three values in 1 dict
        ratings = {rating_type: int(value) for rating in details["Ratings"] for rating_type, value in rating.items()}
        #appending it to the data
        flattened_data.append({
            "review_count": review_number,
            "overall": ratings.get("Overall", None),
            "food": ratings.get("Food", None),
            "service": ratings.get("Service", None),
            "ambience": ratings.get("Ambience", None),
            "date": details["Date"],
            "review": details["Review"]
        })

df_reviews = pd.DataFrame(flattened_data)
df_reviews
```

```python
#count occurance of each overall rating and sort them
overall_dist = df_reviews["overall"].value_counts().sort_index()

#clean dataframe for plotting
overall_dist_df = overall_dist.reset_index()
overall_dist_df.columns = ["overall", "count"]

# Plotting the distribution
sns.barplot(data=overall_dist_df, x="count", y="overall", palette="viridis")
plt.title("Distribution of Overall Ratings")
plt.xlabel("Count")
plt.ylabel("Review Number (Overall Rating)")
plt.tight_layout()
#saving for showing in analysis tab of flask app
plt.savefig("./static/images/overall_rating_distribution.png")
plt.show()
```

```python
with open("analysis.json","r") as f:
    comments_data = json.load(f)

print(len(comments_data))
```

```python
df_comments = pd.DataFrame(comments_data)
df_comments
```

```python
df_comments = df_comments.dropna(axis=1)
```

```python
df_comments
```

```python
#count sentiment values
sentiment_dist = df_comments["Sentiment"].value_counts().sort_index()

#convert series into a data frame
sentiment_dist = sentiment_dist.reset_index()

#rename the columns
sentiment_dist.columns = ["Sentiment", "Count"]

sns.barplot(data=sentiment_dist, x="Sentiment", y="Count", palette="viridis")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.tight_layout()

plt.savefig("static/images/sentiment_distribution_bar.png")

plt.show()
```

```python
#converts date column to string then replace Dined on with empty string and converts the date to datetime objects of pandas
df_reviews["date"] = pd.to_datetime(df_reviews["date"].astype(str).str.replace("Dined on ", "", regex=False))

#groups reviews by date converted into periods of years, gets its count and reset index to create review_count column
popularity_yearly = df_reviews.groupby(df_reviews["date"].dt.to_period("Y")).size().reset_index(name="review_count")

#converting the periods (2024) back to time stamps (2024-01-01 00:00:00)
popularity_yearly["date"] = popularity_yearly["date"].dt.to_timestamp()

plt.figure(figsize=(12, 9))
plt.plot(popularity_yearly["date"], popularity_yearly["review_count"], marker="o", linestyle="-", color="blue")
plt.title("Restaurant Popularity Over Time (Yearly Aggregated)", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Number of Reviews", fontsize=14)
plt.grid(visible=True, linestyle="--", alpha=0.7)
plt.tight_layout()

plt.savefig("static/images/popularity_over_time_yearly.png")
plt.show()
```
